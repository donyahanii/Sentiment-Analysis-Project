```python
# We removed the "Id" column as it’s not useful for the model.
data = data.drop(columns=["Id"])
#group ratings into one category and generate a new dataset
def rating_to_sentiment(rating):
    if rating <= 2:
        return "Negative"
    elif rating == 3:
        return "Neutral"
    else:
        return "Positive"

data["Sentiment"] = data["Rating"].apply(rating_to_sentiment)
```

```python
##Download necessary NLTK resources
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.tokenize import TreebankWordTokenizer
import re
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
tokenizer = TreebankWordTokenizer()
# Function to clean, preprocess, and tokenize text
def preprocess_reviews(text):
    if pd.isnull(text):
        return []
 # Remove special characters (only keep alphabetic characters and spaces)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
 # Convert text to lowercase
    text = text.lower()
# Tokenize the text (split into words)
    tokens = tokenizer.tokenize(text)
# Remove stop words and apply tokenizer
    cleaned_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return cleaned_tokens
```

```python
# Apply the preprocessing function to the 'Review' column and tokenize in one step
data['cleaned_reviews_tokens'] = data['Review'].apply(preprocess_reviews)
data.head()
```

```python
tokenized_sentences = data['cleaned_reviews_tokens'].tolist()

word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=5, workers=4)
```

```python
import numpy as np

def get_review_vector(tokens, model, vector_size=100):
    # Only keep words in the model's vocabulary
    valid_words = [word for word in tokens if word in model.wv.key_to_index]

    if not valid_words:
        return np.zeros(vector_size)  # return zero-vector if no valid words

    word_vectors = [model.wv[word] for word in valid_words]
    return np.mean(word_vectors, axis=0)  # average vector
```

```python
data['review_vector'] = data['cleaned_reviews_tokens'].apply(lambda tokens: get_review_vector(tokens, word2vec_model))
data['review_vector']
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.preprocessing import LabelEncoder
import numpy as np

# 1. Feature vectors
X = np.stack(data['review_vector'].values)

# 2. Sentiment labels (strings)
y_sentiment = data['Sentiment']

# 3. Encode sentiment labels to numeric
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_sentiment)

# 4. Define sampling strategy: oversample only the 'neutral' class
# Find the numeric label for 'neutral'
neutral_label = list(label_encoder.classes_).index('Neutral')

# Let's say you want to oversample 'neutral' to 20000 samples
smote = SMOTE(sampling_strategy={neutral_label: 20000}, random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

# 5. decode numeric labels back to sentiment strings
y_resampled_sentiment = label_encoder.inverse_transform(y_resampled)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# If your labels are still strings ('positive', 'neutral', 'negative'), encode them
le = LabelEncoder()
y_encoded = le.fit_transform(y_resampled)  # 0 = negative, 1 = neutral, 2 = positive

# Define the model
log_model = LogisticRegression(max_iter=3000, random_state=42)

# Define scoring metrics
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision_macro': make_scorer(precision_score, average='macro'),
    'recall_macro': make_scorer(recall_score, average='macro'),
    'f1_macro': make_scorer(f1_score, average='macro')
}

# Perform 5-fold cross-validation
cv_results = cross_validate(log_model, X_resampled, y_encoded, cv=5, scoring=scoring)

# Print average scores
print("Average Cross-Validation Scores:")
print(f"Accuracy: {np.mean(cv_results['test_accuracy']):.4f}")
print(f"Precision (macro): {np.mean(cv_results['test_precision_macro']):.4f}")
print(f"Recall (macro): {np.mean(cv_results['test_recall_macro']):.4f}")
print(f"F1 Score (macro): {np.mean(cv_results['test_f1_macro']):.4f}")

```
